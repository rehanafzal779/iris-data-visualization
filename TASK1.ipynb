```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
iris = sns.load_dataset("iris")
```

```python
# Show shape (rows, columns)
print("Shape of dataset:", iris.shape)

# Show column names
print("\nColumn names:", iris.columns)

# Show first 5 rows
print("\nFirst 5 rows:\n", iris.head())
```

```python
# Scatter plot: Sepal Length vs Sepal Width
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Histogram of sepal_length
sns.histplot(iris["sepal_length"], kde=True, bins=20, color="green")
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot for all numeric columns
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Dataset")
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of petal_length by species
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Petal Length by Species")
plt.show()
```
